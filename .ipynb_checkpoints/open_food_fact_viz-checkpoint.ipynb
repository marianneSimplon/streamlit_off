```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
plot = data.replace(np.nan, 0)
```

## Cration de la colonne "Score saturated fat" 

Cration d'une colonne avec un score cre de la manire suivante :

- **Cration des 4 bins** rpartissant les donnes dans des intervalles.
- **Encoding** des colonnes
- **Moidifcation des valeurs** des colonnes en fonction de leur intervalle (plus grand est la valeur maximale de l'intervalle, plus grand est son score)

- **Fusion des colonnes encodes** en une seule, la colonne "Score saturated fat" (+ suppression des colonnes encodes)

### Cration des 4 bins

```python
plot['Score saturated fat']=pd.qcut(plot['saturated-fat_100g'],duplicates='drop', q=5)
```

```python
plot['Score saturated fat'].unique
```

```python
plot[['Score saturated fat']].info()
```

### Encoding des colonnes

```python
one_hot = pd.get_dummies(plot['Score saturated fat'])
plot = plot.drop('Score saturated fat',axis=1)
```

```python
plot = plot.join(one_hot)
```

### Modification des valeurs

```python
plot=plot.replace({plot.columns[60]: {1: 2}})
plot=plot.replace({plot.columns[61]: {1: 3}})
plot=plot.replace({plot.columns[62]: {1: 4}})

plot = plot.rename(columns={plot.columns[59]:"score 1"})
plot = plot.rename(columns={plot.columns[60]:"score 2"})
plot = plot.rename(columns={plot.columns[61]:"score 3"})
plot = plot.rename(columns={plot.columns[62]:"score 4"})
```

```python
plot.head()
```

### Fusion des colonnes encodes

```python
plot['Score saturated fat']= plot['score 1'] + plot['score 2'] + plot['score 3'] + plot['score 4']
```

```python
plot = plot.drop(columns=['score 1','score 2', 'score 3', 'score 4'], axis = 1)
```

```python
fig = px.histogram(plot, x="Score saturated fat")
fig.show()
```

```python
#px.box(plot,y="Score saturated fat")
```
